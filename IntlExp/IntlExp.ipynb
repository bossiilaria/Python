```python
#My understanding is that fullyvaxDF['TravelOpenness'] selects the first level of the columns and returns a DataFrame that
#is singly-indexed. Then another Python operation selects the series indexed by 'TravelOpenness'. Pandas sees these 
#operations as separate events. e.g. separate calls, so it has to treat them as linear operations, they happen one after 
#another. Instead .loc[:, 'TravelOpenness'] passes a nested tuple of
#(slice(None),('TravelOpenness')) to a single call. This allows pandas to deal with this as a single entity.

#Furthermore this order of operations can be significantly faster, and allows one to index both axes if so desired.

#In addition, it’s very hard to predict whether the fullyvaxDF['TravelOpenness'] will return a view or a copy
#(it depends on the memory layout of the array, about which pandas makes no guarantees), and therefore whether 
#pandas will modify a temporary object that gets thrown out immediately afterward or if it will store it in memory
```
