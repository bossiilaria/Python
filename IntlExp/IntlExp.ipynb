```python
#https://www.youtube.com/watch?v=FDZ9dG6gbUw
import tkinter as tk
```

```python
counter = 0
def counter_label(label):
    counter = 0
    def count():
        global counter
        counter += 1
        label.config(text = str(counter))
        label.after(1000, count)
    count()
```

```python
root = tk.Tk()
```

```python
root.title('counter')
```

```python
#Create Label
label = tk.Label(root, fg = 'dark green')
```

```python
label.pack()
```

```python
counter_label(label)
```

```python
#Create button
button = tk.Button(root, text = 'stop', width = 40, command = root.destroy)
```

```python
button.pack()
```
